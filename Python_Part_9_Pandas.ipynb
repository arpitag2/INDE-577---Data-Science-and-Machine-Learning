```python
def make_target(quality):
    if quality == 5:
        return 0
    elif quality == 6:
        return 1
    else:
        return 2

wine["target"] = list(map(make_target, wine.quality))
wine
```

```python
import seaborn as sns # import seaborn!
sns.set_theme() # Set a gray background theme 

plt.figure(figsize = (10, 8), dpi = 80)

# Grab unique values in the species column
quality_list = list(set(wine.quality))

for color, quality in zip(["magenta", "seagreen"], [5, 6]):
    temp_wine = wine[wine.quality == quality] # Make a DataFrame for a specific species 
    plt.scatter(temp_wine.citric_acid,
                temp_wine.residual_sugar,
                c = color, # Marker color
                s = temp_wine.chlorides * 100, # Marker size 
                alpha = .56, # Marker intensity 
                label = quality # Marker label
                )

plt.xlabel("Citric Acid", fontsize  = 15)
plt.ylabel("Residual Sugar", fontsize = 15)
plt.title("The Wine Data", fontsize = 20)
plt.xlim(4, 7.5)
plt.legend(fontsize = 14)
plt.show()
```
